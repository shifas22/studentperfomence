```python
import pandas as pd
```

```python
df=pd.read_csv(r"Student_Performance.csv")
```

```python
df
```

```python
y=df["Performance Index"]
```

```python
X=df.drop("Performance Index",axis=1)
X
```

```python
cat_data=X.select_dtypes("O")
```

```python
cat_data
```

```python
X.drop("Extracurricular Activities",axis=1,inplace=True)
```

```python
X
```

```python
cat_data
```

## Encoding Using Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
le
```

```python
encoded=le.fit_transform(cat_data)
```

```python
X["ECA"]=encoded
```

```python
X
```

```python
cat_data
```

## SCALING DATA

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler
```

```python
Xscaled=scaler.fit_transform(X)
```

```python
Xscaled
```

## Train Test split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(Xscaled,y,test_size=0.2,random_state=32)
```

```python
X.shape
```

```python
xtrain.shape
```

```python
xtest.shape
```

- The above data gives we have taken 80 % of data for training and rest for testing

## Model Building

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model
```

```python
model.fit(xtrain,ytrain)
```

```python
predict=model.predict(xtest)
```

```python
predict
```

```python
model.coef_
```

- the above generated are the slopes of five features mentioned in our data

```python
model.intercept_
```

## Saving Model

```python
import joblib
```

```python
joblib.dump(le,"Encoder.pkl")
```

```python
joblib.dump(scaler,"Scaler.pkl")
```

```python
joblib.dump(model,"Model.pkl")
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
```

```python
mean_absolute_error(ytest,predict)
```

```python
mean_squared_error(ytest,predict)
```

```python
root_mean_squared_error(ytest,predict)
```

```python
r2_score(ytest,predict)
```
